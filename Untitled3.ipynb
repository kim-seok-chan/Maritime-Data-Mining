```python
# [1] 기본 라이브러리 불러오기
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
uploaded = files.upload()
# [2] 데이터 파일 불러오기 (같은 경로에 csv 파일이 있다고 가정)
file_paths = {
    'USA': 'usa-coviddata.csv',
    'Hawaii': 'hawaii-coviddata.csv',
    'New York': 'ny-coviddata.csv',
    'Los Angeles': 'la-coviddata.csv'
}

dfs = {region: pd.read_csv(path) for region, path in file_paths.items()}

# [3] 인구 수 (2020 기준, 단위: 명)
populations = {
    'USA': 331_000_000,
    'Hawaii': 1_416_000,
    'New York': 19_840_000,
    'Los Angeles': 10_040_000
}

# [4] 인구비례 감염자 수 (인구 10만 명당 신규 확진자 수) 계산
processed = {}
for region, df in dfs.items():
    df['start_date'] = pd.to_datetime(df['start_date'])
    df['cases_per_100k'] = (df['new_cases'] / populations[region]) * 100_000
    df = df[['start_date', 'cases_per_100k']].rename(columns={'cases_per_100k': region})
    df = df.set_index('start_date')
    df = df[~df.index.duplicated(keep='first')]  # 중복 날짜 제거
    processed[region] = df

# [5] 하나의 DataFrame으로 병합
result_df = pd.concat(processed.values(), axis=1).sort_index().dropna().round(2)

# [6] 그래프 시각화
plt.figure(figsize=(14, 6))
for region in result_df.columns:
    plt.plot(result_df.index, result_df[region], label=region)

plt.title('Weekly COVID-19 Cases per 100,000 People (US & Selected States)')
plt.xlabel('Week Start Date')
plt.ylabel('Cases per 100k Population')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# [7] 데이터프레임 확인 (선택)
result_df.tail()
```
